<font size="5">
    $2ln(x) - \frac{x}{2} + 1 = 0$

```python
ox = np.linspace(10**(-5),50,10**5)
def f(x):
    return 2*np.log(x) - x/2 + 1

y = [f(i) for i in ox]

fig, ax = plt.subplots(1, figsize = (14,8))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1));
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1));

ax.axhline(y=0, color='k', linestyle='-')
ax.plot(ox,y)
ax.grid()
```

```python

def bisection(a,b):
    xl = a
    xr = b
    n = 0
    while(np.abs(xl-xr) >= 10**(-14)):
        n += 1
        c = (xl+xr)/2
        prod = f(xl)*f(c)
        if prod > 0:
            xl = c
        else:
            if prod < 0:
                xr = c
    
    return c,n
n1 = 0
n2 = 0
root2,n2 = bisection(10,15)
print(root2,n2)

root1,n1 = bisection(10**(-5),5)
print(root1,n1)
```

```python
fig1, ax1 = plt.subplots(1, figsize = (14,8))
ax1.xaxis.set_minor_locator(ticker.MultipleLocator(1));
ax1.yaxis.set_minor_locator(ticker.MultipleLocator(1));

ax1.axhline(y=0, color='k', linestyle='-')
ax1.axvline(x=root1, color='k', linestyle='-')
ax1.axvline(x=root2, color='k', linestyle='-')

ox = np.linspace(root1*0.5,root2*1.1,10**5)
def f(x):
    return 2*np.log(x) - x/2 + 1

y = [f(i) for i in ox]

ax1.plot(ox,y)

ax1.grid()
```

<font size="5">
    Метод Ньютона<br>
    $f(x) = 2ln(x) - \frac{x}{2} + 1 = 0$<br>
    $f'(x) = \frac{2}{x} - \frac{1}{2}$<br>
    $x_{n+1} = x_{n} - \frac{f(x_{n})}{f'(x_{n})}$<br>

```python
def f(x):
    return 2*np.log(x) - x/2 + 1

def f_prime(x):
    return 2/x - 1/2

def newton(x):
    return x - f(x)/f_prime(x)

root = 7
i = 0
while (abs(f(root)) >= 10**(-14)):
    root = newton(root)
    i+=1
    
    
print("Newton:",root,i)
print("Bisection",root2,n2)
```
